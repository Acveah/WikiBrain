```python
# Create the functions in a matrix
f_len = 100
f_len2 = f_len//2
fmatrix = np.zeros((5,f_len))
fmatrix[1,f_len2:f_len2+10] = 1
fmatrix[2,f_len2+5:f_len2+15] = 1
fmatrix[3,f_len2+10:f_len2+20] = 1
fmatrix[4,10:30] = 1

# Add some random
np.random.seed(1)
fmatrix = fmatrix + np.random.rand(5,f_len) * 0.5
```

```python
for i in range(fmatrix.shape[0]):
    plt.plot(fmatrix[i,:]+i*2)
#plt.axis('off')
#plt.xticks([])
#plt.yticks([])
plt.tick_params(axis='both', which='both', bottom=False,
                top=False, labelbottom=False, right=False, left=False, labelleft=False)
```
