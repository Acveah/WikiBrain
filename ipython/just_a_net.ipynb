```python
import sys

import matplotlib.pyplot as plt
from networkx import nx

n = 7  # 10 nodes
m = 17  # 20 edges

# G = nx.gnm_random_graph(n, m)

# G = nx.powerlaw_cluster_graph(7,3,0.3)
G = nx.generators.directed.random_k_out_graph(7,3,0.5)

# some properties
# print("node degree clustering")
# for v in nx.nodes(G):
#     print('%s %d %f' % (v, nx.degree(G, v), nx.clustering(G, v)))

# print the adjacency list to terminal
# try:
#     nx.write_adjlist(G, sys.stdout)
# except TypeError:  # Python 3.x
#     nx.write_adjlist(G, sys.stdout.buffer)

labels={}
labels[0]=r'Page 1'
labels[1]=r'Page 2'
labels[2]=r'Page 3'
labels[3]=r'Page 4'
labels[4]=r'Page 5'
labels[5]=r'Page 6'
labels[6]=r'Page 7'

pos=nx.spring_layout(G)

pos_higher = {}

for key, value in pos.items():
    pos_higher[key] = (value[0], value[1]+0.07)

plt.figure(figsize=(11,10))
nx.draw_networkx_labels(G,pos_higher,labels,font_size=23,font_weight='bold')

nx.draw(G, pos, node_color='#3090C7')
nx.draw_networkx_edges(G, pos, width=5.0, edge_color='#505050', arrows=True, arrowsize=20)
# plt.savefig("net.svg", format='svg',dpi=200, bbox_inches='tight')
```

```python
import numpy as np
# Create the functions in a matrix
f_len = 100
f_len2 = f_len//2
fmatrix = np.zeros((7,f_len))
fmatrix[1,f_len2:f_len2+10] = 1
fmatrix[2,f_len2+15:f_len2+25] = 1
fmatrix[2,f_len2+45:f_len2+50] = 1
fmatrix[3,f_len2+10:f_len2+20] = 1
fmatrix[4] = np.random.rand(f_len)
fmatrix[5,10:30] = np.random.rand(20)
fmatrix[6,50:70] = np.random.rand(20)

# Add some random
# np.random.seed(1)
fmatrix = fmatrix + np.random.rand(7,f_len) * 0.7
```
